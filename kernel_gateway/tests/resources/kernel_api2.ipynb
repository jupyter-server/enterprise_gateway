# API Creation 
This notebook is a sample of how to author a REST API in the notebook environment.

```python
import json
```

```python
hello_message = 'hello {}'
people = ['Corey', 'Nitin', 'Pete']
```

## Hello

```python
# GET /hello
print(hello_message.format('world'))
```

```python
REQUEST = json.dumps({
    'args': { 
        'person' : people
    }
})
```

```python
# GET /hello/person
req = json.loads(REQUEST)
hello_person = req['args']['person'][0]
print(hello_message.format(hello_person))
```

```python
REQUEST = json.dumps({
    'args': { 
        'person' : people
    }
})
```

```python
# GET /hello/persons
req = json.loads(REQUEST)
hello_persons = req['args']['person']
print(hello_message.format(', '.join(hello_persons)))
```

```python
# GET /hello/people
hello_people = people
print(hello_message.format(', '.join(hello_people)))
```

```python
REQUEST = json.dumps({
    'path' : {
        'person' : 'test_person'
    }
})
```

```python
# GET /hello/:person
req = json.loads(REQUEST)
print(hello_message.format(req['path']['person']))
```

## Messages

```python
# GET /message
print(hello_message)
```

```python
REQUEST = json.dumps({
    'body' : 'test value'
})
```

```python
# PUT /message
req = json.loads(REQUEST)
hello_message = req['body']
print(hello_message)
```

## People

```python
# GET /people
print(json.dumps(people))
```

```python
REQUEST = json.dumps({
    'body' : ['Rick', 'Maggie', 'Glenn', 'Carol', 'Daryl']
})
```

```python
# POST /people
req = json.loads(REQUEST)
people = req['body']
print(json.dumps(people))
```

```python
REQUEST = json.dumps({
    'body' : 'Michonne'
})
```

```python
# PUT /people
req = json.loads(REQUEST)
people.append(req['body'])
print(json.dumps(people))
```

```python
REQUEST = json.dumps({
    'path' : {
        'index' : 1
    }
})
```

```python
# DELETE /people/:index
req = json.loads(REQUEST)
people.remove(people[int(req['path']['index'])])
print(json.dumps(people))
```

```python
# GET /sleep/:time
req = json.loads(REQUEST)
from time import sleep
sleep_time = int(req['path']['time'])
sleep(sleep_time)
print("Slept for {} seconds".format(sleep_time))
```
